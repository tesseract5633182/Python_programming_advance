<b>Q.1 You are given two strings s and t. String t is generated by randomly
shuffling string s and then adding one more letter at a random position.
Return the letter that was added to t.

```python
def find_the_difference(string1, string2):
    dict_count = {}
    for i in string2:
        if i in dict_count:
            dict_count[i] += 1
        else:
            dict_count[i] = 1
    for i in string1:
        dict_count[i] -= 1
    for h1 in dict_count:
        if dict_count[h1] == 1:
            return h1
 
```

```python
print(find_the_difference("abcd",'atbcd'))
print(find_the_difference("abcd",'abcde'))
```

many data types are in those arguments. Finally return the total in a list.

```python
def count_datatypes(lst):
    try:
        data_type = []
        data_type_count = {}
        count =0 
        data_types = ['int', 'str', 'bool', 'list', 'tuple', 'dictionary']
        output = []
        
        for i in lst:
            data_type.append(type(i).__name__)
        for j in set(data_type):
            for k in (data_type):
                if(j==k):
                    count +=1
            data_type_count[j]=count
            count = 0
        for i in data_types:
            try:
                output.append(data_type_count[i])
            except Exception as e:
                output.append(0)
        return output
    except Exception as e:
        return e
```

```python
print(count_datatypes([1,45,'Hi',False]))
print(count_datatypes([[10,20],('t','OK'),2,3,1]))
```

<b>Q.3 A Fibonacci string is a precedence of the Fibonacci series. It works with
any two characters of the English alphabet (as opposed to the numbers 0 and
1 in the Fibonacci series) as the initial items and concatenates them together
as it progresses in a similar fashion as the Fibonacci series.

```python
def fib_str(k, lst):
    fib_ = lst
    try:
        for i in range(0,k-2):
            fib_.append(fib_[-1]+fib_[-2])
        return ", ".join(fib_)    
    except Exception as e:
        return e
```

```python
print(fib_str(3,['j','h']))
print(fib_str(5,['e','a']))
```

<b>Q.5 The Fibonacci sequence is a classic use case for recursive functions since
the value of the sequence at a given index is dependent on the sum of the
last two values. However, the recursion tree created by solving the Fibonacci
sequence recursively can grow quite fast. Therefore it can be important to
think about the implications of running a function recursively. Depending on
the size of n needed and the capabilities of the system in question you might
want to take a different approach.
Write a non-recursive function that takes an integer n and returns the
Fibonacci sequence&#39;s value at index n.

```python
def fib(n):
    fibonacci=(0,1)
    for i in range(2,n+1):
        fibonacci+=(fibonacci[i-2]+fibonacci[i-1],)
    return fibonacci[-1]
```

```python
print(fib(6))
print(fib(2))
```
